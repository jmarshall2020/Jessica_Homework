Pymaceuticals specializes in drug-based, anti-cancer pharmaceuticals. In their most recent efforts, they've since begun screening for potential treatments to squamous cell carcinoma (SCC), a commonly occurring form of skin cancer.

As their Chief Data Analyst, you've been given access to the complete data from their most recent animal study. In this study, 250 mice were treated through a variety of drug regimes over the course of 45 days. Their physiological responses were then monitored over the course of that time. Your objective is to analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare.

To do this you are tasked with:

* Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
* Creating a scatter plot that shows how the number of [metastatic](https://en.wikipedia.org/wiki/Metastasis) (cancer spreading) sites changes over time for each treatment.
* Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)
* Creating a bar graph that compares the total % tumor volume change for each drug across the full 45 days.
* Include 3 observations about the results of the study. Use the visualizations you generated from the study data as the basis for your observations.

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
clinical_data_complete = pd.merge(clinical_trial_df, mouse_drug_df, how="left", on=["Mouse ID","Mouse ID"])

# Display the data table for preview

clinical_data_complete.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_vol_mean = clinical_data_complete.groupby(['Drug', 'Timepoint']).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_mean_df = pd.DataFrame(tumor_vol_mean)

#Flatten to column headings
#tumor_vol_mean = tumor_vol_mean.reset_index()

# Preview DataFrame
tumor_mean_df.head().reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#tumor_vol_sem = clinical_data_complete.groupby(["Drug", "Timepoint", "Tumor Volume(mm3)"]).sem()['Tumor Volume(mm3)]']
#d.DataFrame.sem(tumor_mean_df)

# Convert to DataFrame
#tumor_vol_sem_df = pd.DataFrame(tumor_vol_sem)

# Preview DataFrame
#umor_vol_sem_df.head().reset_index()

#tumor_vol_sem.head().reset_index()

# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_vol_sem = clinical_data_complete.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)']

tumor_sem_df = pd.DataFrame(tumor_vol_sem)

tumor_sem_df.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_vol_mean = tumor_vol_mean.reset_index()
tumor_vol_mean_pivot = tumor_vol_mean.pivot(index="Timepoint", columns="Drug")['Tumor Volume (mm3)']

# Preview that Reformatting worked
tumor_vol_sem = tumor_vol_sem.reset_index()

tumor_vol_mean_pivot.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_vol_mean = tumor_vol_mean.reset_index()
tumor_vol_mean_pivot = tumor_vol_mean.pivot(index="Timepoint", columns="Drug")['Tumor Volume (mm3)']

# Preview that Reformatting worked
tumor_vol_sem = tumor_vol_sem.reset_index()
tumor_vol_sem_pivot = tumor_vol_sem.pivot(index="Timepoint", columns="Drug")['Tumor Volume (mm3)']

tumor_vol_sem_pivot.head()
```

```python
#Your objective is to analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo)
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_vol_mean_pivot.index,tumor_vol_mean_pivot['Capomulin'], yerr=tumor_vol_sem_pivot["Capomulin"], color="r",marker="o", markersize=5, linestyle="dashed", linewidth=0.50)


plt.errorbar(tumor_vol_mean_pivot.index,tumor_vol_mean_pivot['Infubinol'], yerr=tumor_vol_sem_pivot["Infubinol"], color="b",marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vol_mean_pivot.index,tumor_vol_mean_pivot['Ketapril'], yerr=tumor_vol_sem_pivot["Ketapril"], color="g",marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_vol_mean_pivot.index,tumor_vol_mean_pivot['Placebo'], yerr=tumor_vol_sem_pivot["Placebo"], color="k",marker="o", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig('analysis/Fig1.png')

plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rates = clinical_data_complete.groupby(["Drug", "Timepoint"]).count()["Tumor Volume (mm3)"]

# Convert to DataFrame
survival_rates_df = pd.DataFrame({"Mouse Count": survival_rates})

# Preview DataFrame
survival_rates_df.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_rates_df = survival_rates_df.reset_index()

pivot_survival_rates_df = survival_rates_df.pivot(index="Timepoint", columns="Drug")["Mouse Count"]
pivot_survival_rates_df


# Preview the Data Frame
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(100 * pivot_survival_rates_df["Capomulin"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * pivot_survival_rates_df["Infubinol"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * pivot_survival_rates_df["Ketapril"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * pivot_survival_rates_df["Placebo"] / 25, "ro", linestyle="dashed", markersize=5, linewidth=0.50)
plt.title("Survival During Treatment")
plt.ylabel("YLabel")
plt.xlabel("XLabel")
plt.grid(axis = "y")
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig('analysis/Fig3.png')

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
drug_mean_percent_change =  100 * (tumor_vol_mean_pivot.iloc[-1] - tumor_vol_mean_pivot.iloc[0]) / tumor_vol_mean_pivot.iloc[0]

#drug_sem_percent_change =  100 * (pivot_sem_metastatic_df.iloc[-1] - pivot_sem_metastatic_df.iloc[0]) / pivot_sem_metastatic_df.iloc[0]

# Display the data to confirm

drug_mean_percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_tuple = (drug_mean_percent_change["Capomulin"],
              drug_mean_percent_change["Infubinol"],
              drug_mean_percent_change["Ketapril"],
              drug_mean_percent_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
sliced_data = np.arange(len(percent_tuple))
width = 1
recent_Pass = ax.bar(sliced_data[0], percent_tuple[0], width, color='green')
recent_Fail = ax.bar(sliced_data[1:], percent_tuple[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel("%Tumor Volume Change")
ax.set_title("Tumor Change Over 45 Day Treatment")
ax.set_xticks(sliced_data + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.grid(True)

# Use functions to label the percentages of changes
def Fail_Label(rects):
   for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., 3,
               '%d%%' % int(height),
               ha='center', va='bottom', color="white")
def Pass_Label(rects):
   for rect in rects:
       height = rect.get_height()
       ax.text(rect.get_x() + rect.get_width()/2., -8,
               '-%d%% ' % int(height),
               ha='center', va='bottom', color="white")
# Call functions to implement the function calls
Pass_Label(recent_Pass)
Fail_Label(recent_Fail)



# Save the Figure
plt.savefig('analysis/Fig4.png')

# Show the Figure
fig.show()
```
